## Hamiltonian of a bulk antiferromagnet on a 2d square lattice

```julia
include("kpaths.jl")
```

Initialize the system parameters

```julia
Rs = ([1.0, 0.0], [0, 1.0])                 # Lattice vectors, can be 2d or 3d
high_sym_line = [:Γ, :M, :X, :Γ]            # Custom k-path over high symmetry momenta
sgnum = 10                                  # point group (10 = 2d square lattice)
N = 100                            # points in the k-mesh
```

Compute `N` k-points in the path defined by `high_sym_line`

```julia
kp = kpath(Rs, sgnum, num_points);          # kpoints
```

Using CairoMakie we can plot the WS cell, and the lines

```julia
plot_kpath(Rs, sgnum, num_points, high_sym_line)
```

If there is no high-symmetry line is passed as an argument it computes the default

```julia
plot_kpath(Rs, sgnum, num_points)
```

```julia
plot_kpath(Rs, sgnum, num_points,  [:Γ,:M])
```

Observable evaluated in the k-line

```julia
obs(x) = x[1]^2 + x[2]^2
plot_observable_in_kpath(obs, Rs, sgnum, num_points, high_sym_line)
```

Example: Dispersion of an antiferromagnet

```julia
# --- Parameters ---
t   = 1.0
tp  = -0.2
tpp = 0.05
μ   = 0.0
Δ   = 2        # AF mean-field gap

# --- Dispersion function ---
ε(k; kws...) = ε(k[1],k[2]; kws...) 
ε(kx, ky; t=t, tp=tp, tpp=tpp, μ=μ) =
    -2t*(cos(kx)+cos(ky)) -
    4tp*cos(kx)*cos(ky) -
    2tpp*(cos(2kx)+cos(2ky)) - μ;
```

Compute the spectrum along the default symmetry line

```julia
plot_observable_in_kpath(ε, Rs, sgnum, num_points)
```

And over a custom line

```julia
plot_observable_in_kpath(ε, Rs, sgnum, num_points, [:Γ, :X, :M,:Γ, :M, :X,:Γ])
```
